```python
fig, ax = plt.subplots(figsize=(8, 6))

sns.set_style("whitegrid")
sns.set_context("talk")
sns.histplot(data=df, x='AGE', hue='PATIENT_TYPE', multiple="stack", palette="pastel", ax=ax)
ax.set_title('Relation Age and Type of Patient')
ax.set_xlabel('Age')
ax.set_ylabel('Count')
ax.legend(title='Hospitalization', loc='upper right', labels=['Yes', 'No'])

plt.show()
```
